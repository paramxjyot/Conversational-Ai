```python
from langchain.embeddings import GPT4AllEmbeddings
from langchain.vectorstores import FAISS

gpt4all_embd = GPT4AllEmbeddings()
vectro_db = FAISS.from_documents(documents = doc , embedding = gpt4all_embd)
```

```python
query_result = gpt4all_embd.embed_query("what is the cost of the course?")
query_result
```

```python
retriever = vectro_db.as_retriever()
```

```python
rdocs = retriever.get_relevant_documents("How long is this course valid")
rdocs
```

```python
from langchain.chains import RetrievalQA

chain = RetrievalQA.from_chain_type(llm=llm,
                            chain_type="stuff",
                            retriever=retriever,
                            input_key="query",
                            return_source_documents=True,
                            )
```

```python
chain("whatis the fees and duration of the course")
```
